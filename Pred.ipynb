```python
import pandas as pd
data=pd.read_csv('time_series_covid19_deaths_global.csv')
data
```

```python
country=list(data['Country/Region'].unique())
death=data.groupby(data['Country/Region']).sum()
death=death.iloc[:,2:]
death_toll=death.sum(axis=0)
```

```python
df=pd.read_csv('time_series_covid19_recovered_global.csv')
df
```

```python
country_rec=list(data['Country/Region'].unique())
rec=df.groupby(data['Country/Region']).sum()
rec=rec.iloc[:,2:]
recno=rec.sum(axis=0)
```

```python
new=pd.read_csv('time_series_covid19_confirmed_global.csv')
```

```python
country_rec=list(data['Country/Region'].unique())
cases=new.groupby(data['Country/Region']).sum()
cases=cases.iloc[:,2:]
caseno=cases.sum(axis=0)
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(50,50)
plt.rcParams.update({'font.size':40})

plt.plot(death_toll,marker='.',color='red',markerfacecolor='black',markersize=10,markeredgewidth=10,label='death toll')
plt.plot(recno,marker='.',color='magenta',markerfacecolor='black',markersize=10,markeredgewidth=10,label='recovered cases')
plt.plot(caseno,marker='.',color='orange',markerfacecolor='black',markersize=10,markeredgewidth=10,label='total cases')
plt.xticks(np.arange(0,75,5),rotation=45)
plt.title('COVID-19 cases vs death toll vs recovered',fontsize=50)
plt.legend(fontsize='x-large')
plt.show()
```

# predictions

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree = 5)
```

```python
Y_india=np.array(active_cases_india).reshape(-1,1)
Y_usa=np.array(active_cases_usa).reshape(-1,1)
Y_spain=np.array(active_cases_spain).reshape(-1,1)
Y_france=np.array(active_cases_france).reshape(-1,1)
Y_china=np.array(active_cases_china).reshape(-1,1)
Y_italy=np.array(active_cases_italy).reshape(-1,1)
```

```python
X_india=np.arange(len(Y_india)).reshape(-1,1)
X_usa=np.arange(len(Y_usa)).reshape(-1,1)
X_italy=np.arange(len(Y_italy)).reshape(-1,1)
X_china=np.arange(len(Y_china)).reshape(-1,1)
X_france=np.arange(len(Y_france)).reshape(-1,1)
X_spain=np.arange(len(Y_spain)).reshape(-1,1)
```

```python

X_poly_india=poly_reg.fit_transform(X_india)
X_poly_usa=poly_reg.fit_transform(X_usa)
X_poly_italy=poly_reg.fit_transform(X_italy)
X_poly_china=poly_reg.fit_transform(X_china)
X_poly_france=poly_reg.fit_transform(X_france)
X_poly_spain=poly_reg.fit_transform(X_spain)

lin_reg2_india=LinearRegression()
lin_reg2_usa=LinearRegression()
lin_reg2_italy=LinearRegression()
lin_reg2_china=LinearRegression()
lin_reg2_france=LinearRegression()
lin_reg2_spain=LinearRegression()

lin_reg2_india.fit(X_poly_india,Y_india)
lin_reg2_usa.fit(X_poly_usa,Y_usa)
lin_reg2_italy.fit(X_poly_italy,Y_italy)
lin_reg2_china.fit(X_poly_china,Y_china)
lin_reg2_france.fit(X_poly_france,Y_france)
lin_reg2_spain.fit(X_poly_spain,Y_spain)
```

```python
pred=np.arange(120).reshape(-1,1)
plt.plot(lin_reg2_india.predict(poly_reg.fit_transform(pred)),color='red',marker='o',markerfacecolor='black',markersize=10,markeredgewidth=10,label=' INDIA predicted cases(poly)')
plt.plot(lin_reg2_china.predict(poly_reg.fit_transform(pred)),color='orange',marker='o',markerfacecolor='black',markersize=10,markeredgewidth=10,label='CHINA predicted cases(poly)')
plt.plot(lin_reg2_usa.predict(poly_reg.fit_transform(pred)),color='magenta',marker='o',markerfacecolor='black',markersize=10,markeredgewidth=10,label='USA predicted cases(poly)')
plt.plot(lin_reg2_italy.predict(poly_reg.fit_transform(pred)),color='blue',marker='o',markerfacecolor='black',markersize=10,markeredgewidth=10,label='ITALY predicted cases(poly)')
plt.plot(lin_reg2_france.predict(poly_reg.fit_transform(pred)),color='black',marker='o',markerfacecolor='black',markersize=10,markeredgewidth=10,label='FRANCE predicted cases(poly)')
plt.plot(lin_reg2_spain.predict(poly_reg.fit_transform(pred)),color='green',marker='o',markerfacecolor='black',markersize=10,markeredgewidth=10,label='SPAIN predicted cases(poly)')

plt.xticks(np.arange(0,120,5),rotation=45)
plt.title('active cases in millions across countries till 150 days',fontsize=50)
plt.ylabel('Affected people in millions',fontsize=50)
plt.xlabel('Dates',fontsize=50)
plt.legend(fontsize='x-large')
plt.show()
```
